#### first day on openinsider is 7/28/2003. start scraping at 8/1/2003. or start at beginning of 2004

```python
import datetime

# year = '2021'
monthDelims = []

def last_day_of_month(any_day):
    next_month = any_day.replace(day=28) + datetime.timedelta(days=4)
    return next_month - datetime.timedelta(days=next_month.day)

for year in range(2022, 2023):
    for month in range(1, 5):
        firstDayString = str(month)+'/01/'+str(year)
        if month<10:
            firstDayString='0'+firstDayString
    #     monthDelims.append(last_day_of_month(datetime.date(year, month, 1)).strftime("%m/%d/%Y"))
        monthDelims.append([firstDayString, last_day_of_month(datetime.date(int(year), month, 1)).strftime("%m/%d/%Y")])
monthDelims
```

### Now making many scrapes broken up by month, then iterating through the pages for that given month and reading into a dataframe

```python
# URL = 'http://openinsider.com/screener?s=&o=&pl=&ph=&ll=&lh=&fd=0&fdr=&td=-1&tdr=01%2F01%2F2021+-+01%2F31%2F2021&fdlyl=&fdlyh=&daysago=&xp=1&xs=1&vl=&vh=&ocl=&och=&sic1=-1&sicl=100&sich=9999&grp=0&nfl=&nfh=&nil=&nih=&nol=&noh=&v2l=&v2h=&oc2l=&oc2h=&sortcol=1&cnt=1000&page=1'
# page = '1'
df = pd.DataFrame(columns=['Trade Date', 'Ticker', 'Insider Name', 'Company Name', 'Title', 'Trade Type', 'Price', 'Qty', 'Owned', 'deltOwned', 'Value'])

for monthDel in monthDelims:
    print()
    print(monthDel)
    for page in range(1, 10):   #should be to 10.  #break out of this if we scrape and it doesn't have "<h3>Filing&nbsp;Date</h3>"
        print('Page Number: ', page)
        URL = 'http://openinsider.com/screener?s=&o=&pl=&ph=&ll=&lh=&fd=0&fdr=&td=-1&tdr=' + (monthDel[0][:2]+'%2F') + (monthDel[0][3:5] + '%2F') + monthDel[0][6:10] + '+-+' + (monthDel[1][:2]+'%2F') + (monthDel[1][3:5] + '%2F') + monthDel[1][6:10] +'&fdlyl=&fdlyh=&daysago=&xp=1&xs=1&vl=&vh=&ocl=&och=&sic1=-1&sicl=100&sich=9999&grp=0&nfl=&nfh=&nil=&nih=&nol=&noh=&v2l=&v2h=&oc2l=&oc2h=&sortcol=1&cnt=1000&page='+str(page)
        page = requests.get(URL, headers=headers)
        
        if "<h3>Filing" not in page.text:
            break
        
        stuff = page.text[(page.text).index('<th><h3>6m</h3></th>')+len('<th><h3>6m</h3></th>'):]
        
#         print(stuff)
        count = 0
        while '_blank' in stuff:
            count +=1
            stuff = stuff[stuff.index('_blank')+8:]
            start = stuff.index('</b></td><td><a href=')+len('</b></td><td><a href=')+2
            tickerEnd = start + stuff[start:].index('>')-1
            compNameStart = tickerEnd + 2
            compNameEnd = compNameStart + stuff[compNameStart:].index('<')
            insiderStart = compNameEnd + stuff[compNameEnd:].index('">') + 2
            insiderEnd = insiderStart + stuff[insiderStart:].index('<')
            titleStart = insiderEnd + len('</a></td><td>')
            titleEnd = titleStart + stuff[titleStart:].index('<')
            typeStart = titleEnd + len('</td><td>')
            typeEnd = typeStart + stuff[typeStart:].index('<')
            priceStart = typeEnd + len('</td><td align=right>$')
            priceEnd = priceStart + stuff[priceStart:].index('<')
            qtyStart = priceEnd + len('</td><td align=right>')
            qtyEnd = qtyStart + stuff[qtyStart:].index('<')
            ownedStart = qtyEnd + len('</td><td align=right>')
            ownedEnd = ownedStart + stuff[ownedStart:].index('<')
            pctChangeStart = ownedEnd + len('</td><td align=right>')
            try:
                pctChangeEnd = pctChangeStart + stuff[pctChangeStart:].index('%')
            except:
                pctChangeEnd = pctChangeStart + stuff[pctChangeStart:].index('<')
    
            
            valueStart = pctChangeEnd + len('%</td><td align=right>')
            valueEnd = valueStart + stuff[valueStart:].index('<')

            df = df.append({'Trade Date': stuff[stuff.index('/td><td align=right><div>')+len('/td><td align=right><div>'):stuff.index('/td><td align=right><div>')+len('/td><td align=right><div>') +10],
                            'Ticker': stuff[start : tickerEnd],
                            'Company Name': stuff[compNameStart : compNameEnd], 
                            'Insider Name' : stuff[insiderStart : insiderEnd],
                            'Title': stuff[titleStart : titleEnd],
                            'Trade Type': stuff[typeStart : typeEnd],
                            'Price': stuff[priceStart : priceEnd],
                            'Qty': stuff[qtyStart : qtyEnd],
                            'Owned': stuff[ownedStart : ownedEnd],
                            'deltOwned': stuff[pctChangeStart : pctChangeEnd],
                            'Value': stuff[valueStart:valueEnd] }, 
                           ignore_index=True)
    
    
```
